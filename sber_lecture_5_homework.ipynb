#### ЦПТ для различных распределений

Возьмите то непрерывное распределение, которое вам больше всего понравится из списка доступных в модуле scipy.stats.

Сгенерируйте примеры из распределения

```python
arcsine_rv = sts.arcsine()
real_mean, real_var = arcsine_rv.mean(), arcsine_rv.var() # настоящее среднее и настоящая дисперсия случайной величины
```

```python
def create_rv_means_sample(sample_size:int, n:int) -> pd.Series:
    rv = sts.arcsine()
    sample = pd.Series([rv.rvs(n).mean() for _ in range(sample_size)])
    return sample
```

Постройте гистограмму выборки и теоретическую плотность распределения

```python
s = arcsine_rv.rvs(50)
plt.hist(s, alpha=.3)

theory_rv = sts.norm(loc=real_mean, scale=real_var)
x = np.linspace(0, 1, 20)
pdf = theory_rv.pdf(x)
plt.plot(x, pdf, label=f"Теоретическое распределение ({n})", alpha=.5)

plt.show()
```

```python
sample_size = 50
n = 10
sample = create_rv_means_sample(sample_size, n)
plt.hist(sample, alpha=.3)

theory_rv = sts.norm(loc=real_mean, scale=real_var)
x = np.linspace(0, 1, 20)
pdf = theory_rv.pdf(x)
plt.plot(x, pdf, label=f"Теоретическое распределение ({n})", alpha=.5)

plt.legend()
plt.show()
```
